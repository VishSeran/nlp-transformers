# Train a word2vec model from gensim

Here's a simple hands-on exercise to train a word2vec model using `gensim` library.
In this example, you have a small corpus consisting of four sentences. 

### Prepare your corpus:


```python
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
```

```python
sentences = [["The", "dog", "loves", "to", "run"],
             ["Running", "is", "great", "for", "health"],
             ["The", "cat", "enjoys", "sleeping"],
             ["Cats", "and", "dogs", "are", "popular", "pets"],
             ["He", "drinks", "coffee", "every", "morning"],
             ["Coffee", "keeps", "me", "awake", "and", "focused"],
             ["She", "loves", "reading", "books", "at", "night"],
             ["Books", "are", "a", "great", "source", "of", "knowledge"]]
```

```python
sentences_lower = [[word.lower() for word in sentence]for sentence in sentences]
```

The `size` parameter specifies the dimensionality of the word embeddings (in this case, 100). The `window` parameter determines the size of the context window. The `min_count` parameter sets the minimum frequency of a word to be included in the training process. Finally, the `workers` parameter controls the number of threads used for training.


```python
word2vec_model = Word2Vec(sentences_lower,vector_size=100,window=2,min_count=1,workers=4)
```

```python
# create vocabulary
word2vec_model.build_vocab(sentences_lower,progress_per=10000)
```

Model Training

```python
word2vec_model.train(sentences_lower,total_examples=word2vec_model.corpus_count,epochs=10, report_delay=1)
```

```python
# find the similar word
similar_word = word2vec_model.wv.most_similar("dog")
print("similar words to readings:", similar_word)
```

```python
# calculating words similarities
similarities = word2vec_model.wv.similarity("dog", "she")
print("Similarity between 'dog' and 'cat':", similarities)
```
