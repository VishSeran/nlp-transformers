# Converting Words into Features

* Initially the basic idea behind this is process the natural lanuage and feed the text format to the machince as a numeric type that machine can understand and process.

* We use One-Hot Encoding for dataset to convert text based data into numeric type and process that converted data using a neural network.

* By this method a lot of data have been feed as input for neural network, that have cost the computational power more and huge time to give a output.

* Because of these reasons, we follow a different method. Instead of feed the one-hot encoded data into neural network, we feed the <b>token index</b> of the data as the input of the neural network.

```python
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator  
```

```python
corpus = [
    "I love this product it is amazing",
    "This is the worst thing I ever bought",
    "Absolutely fantastic experience",
    "I hate this so much"
]
```

### Build Tokrnizer and Vocabulary

```python
tokenizer = get_tokenizer("spacy", language="en_core_web_sm")
```

```python
# define yield tokenizer function
def yield_tokens (data_iterate):
    for data in data_iterate:
        yield tokenizer(data)
```

```python
# define iterator
data_iterate = iter(corpus)
```

```python
# define vocab
vocan = build_vocab_from_iterator(yield_tokens(data_iterate))
```
